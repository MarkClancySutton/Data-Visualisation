```r
#This code was geenrated through ai software 

# Load necessary libraries
library(ggplot2)
library(dplyr)

# Load the mtcars dataset
data(mtcars)

# Categorize mpg variable
mtcars$mpg_category <- cut(mtcars$mpg, breaks = c(10, 20, 30,40), labels = c("10-20", "20-30","30-40"))






# Custom Colors and Labels
pie_chart_version2 <- ggplot(mtcars, aes(x = 1, fill = factor(mpg_category))) +
  geom_bar(width = 1, color = "white") +
  geom_text(stat = "count", aes(label = paste(round((..count..)/sum(..count..) * 100, 1), "%")), position = position_stack(vjust = 0.5), color = "white") +
  coord_polar(theta = "y") +
  labs(title = "MPG Category Distribution", subtitle = "Custom Colors and Labels", fill = "") +
  theme_void()



# Define custom colors
custom_colors <- c("#3498db", "#e74c3c", "#2ecc71", "#f39c12", "#9b59b6", "#34495e")




# Bar chart for mpg values with larger and more readable text
bar_chart_mpg <- ggplot(mtcars, aes(x = rownames(mtcars), y = mpg)) +
  geom_bar(stat = "identity", fill = "#3498db", color = "#2c3e50", alpha = 0.8) +
  geom_text(aes(label = round(mpg, 1)), vjust = -0.5, size = 3, color = "#e74c3c") +
  labs(title = "MPG Values for Different Car Models", x = "Car Model", y = "Miles per Gallon (mpg)") +
  theme_minimal() +
  theme(
    axis.text.x = element_text(angle = 45, hjust = 1, vjust = 1, size = 8),  # Adjust text size
    axis.title = element_text(size = 12),
    plot.title = element_text(size = 16, hjust = 0.5)
  )


# Display the bar chart
print(bar_chart_mpg)








```

```r

# Bar chart
bar_chart <- ggplot(mtcars, aes(x = mpg_category)) +
  geom_bar(fill = "skyblue", color = "black") +
  labs(title = "MPG Category Distribution", x = "MPG Category", y = "How many ") +
  theme_minimal()
# Display the visualizations
print(bar_chart)
```

```r
# Pie chart with percentages
pie_chart <- ggplot(mtcars, aes(x = 1, fill = mpg_category)) +
  geom_bar(width = 1) +
  geom_text(stat = "count", aes(label = paste(round((..count..)/sum(..count..) * 100, 1), "%")), position = position_stack(vjust = 0.5)) +
  coord_polar(theta = "y") +
  labs(title = "MPG Category Distribution with Percentages") +
  theme_void()

print(pie_chart)
```

```r
# Classic Pie Chart with Custom Wedge Colors
pie_chart_version1_custom_colors <- ggplot(mtcars, aes(x = 1, fill = factor(mpg_category))) +
  geom_bar(width = 1) +
  geom_text(stat = "count", aes(label = paste(round((..count..)/sum(..count..) * 100, 1), "%")), position = position_stack(vjust = 0.5), color = "white") +
  coord_polar(theta = "y") +
  labs(title = "MPG Category Distribution", subtitle = "Classic Pie Chart with Custom Colors", fill = "") +
  scale_fill_manual(values = custom_colors) +  # Set custom colors
  theme_void()
# Display the classic pie chart with custom colors
print(pie_chart_version1_custom_colors)
```

```r
# Simplified Pie Chart
pie_chart_version3 <- ggplot(mtcars, aes(x = 1, fill = factor(mpg_category))) +
  geom_bar(width = 1) +
  coord_polar(theta = "y") +
  labs(title = "MPG Category Distribution", subtitle = "Simplified Pie Chart", fill = "") +
  theme_void()
# Display the simplified pie chart
print(pie_chart_version3)
```

```r
# Histogram
histogram <- ggplot(mtcars, aes(x = mpg, fill = mpg_category)) +
  geom_histogram(binwidth = 5, color = "black", position = "identity") +
  labs(title = "MPG Distribution", x = "MPG", y = "Count") +
  theme_minimal()
print(histogram)
```
